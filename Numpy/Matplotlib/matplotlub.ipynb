# Bar Graph Testing


```python
import numpy as np

labels = ['G1','G2','G3','G4','G5']
blue_means = [20,34,30,35,27]
red_means= [25,32,34,20,25]
x = np.arange(len(labels))
width = .35
fig, ax = plt.subplots()
rects1 = ax.bar(x-width/2,blue_means,width,label= 'Blue')
rects2 = ax.bar(x+width/2,red_means,width,label= 'Red')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and color')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
fig.tight_layout()
plt.show()
```
